```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
iris['species'].nunique()
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
```

```python
g.map(plt.scatter)
```

```python
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
g=sns.FacetGrid(data=tips   , row = 'time' , col ='smoker' )
g.map(sns.distplot , 'total_bill' , kde=False)
```

```python
g=sns.FacetGrid(data=tips , row='time' , col='sex')
g.map(sns.distplot , 'total_bill' , kde=False)
```
