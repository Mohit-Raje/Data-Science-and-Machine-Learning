```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex' , y='total_bill' , data=tips)
```

```python
sns.barplot(x='sex' , y='total_bill' , data=tips , estimator=np.std)
```

```python
sns.countplot(x='sex' , data=tips)
```

```python
sns.boxplot(x='day' , y='total_bill' , data=tips)
```

```python
sns.boxplot(x='day' ,y='total_bill' , data=tips , hue='smoker')
```

```python
sns.boxplot(x='sex' , y='total_bill' , data=tips )
```

```python
sns.violinplot(x='day' , y='total_bill' , data=tips)
```

```python
sns.violinplot(x='day' , y='total_bill' , data=tips , hue='sex' , split=True)
```

```python
sns.stripplot(x='day' , y='total_bill' , data=tips)
```

```python
sns.stripplot(x='day' , y='total_bill' , data=tips , hue='sex' ,split=True)
plt.legend(loc='best')
```

```python
sns.swarmplot(x='day' , y='total_bill' , data=tips)
```

```python
sns.violinplot(x='day' , y='total_bill' , data=tips)
sns.swarmplot(x='day' , y='total_bill' , data=tips , color='black')
```

```python
sns.factorplot(x='day' , y='total_bill' , data=tips , kind='bar')
```

```python
sns.factorplot(x='day' , y='total_bill' , data=tips , kind='violin')
```

```python
sns.factorplot(x='day' , y='total_bill' , data=tips , kind='box')
```

```python
sns.factorplot(x='day' , y='total_bill' , data=tips , kind='swarm')
```

```python
sns.violinplot(x='day' , y='total_bill' , data=tips , hue='time' , spilt=True)
plt.legend(loc='upper left')
```
