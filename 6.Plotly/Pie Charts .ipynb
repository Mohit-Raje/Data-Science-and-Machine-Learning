```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Simple Pie chart

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)
l=['Apple' , 'Samsung' , 'oneplus' , 'Vivo' , 'Oppo']
s=[34.5 , 50.34 , 12.4  , 20.45 , 30.45]
plt.pie(x=s, labels=l , autopct='%.2f%%')
plt.legend(loc='upper left')
```

## Exploded Pie chart

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(10)

explode=(0.1,0,0,0,0)
plt.pie(x=s, labels=l , explode=explode , autopct="%.2f%%" ,shadow=True )
plt.legend(loc='upper left')
```

## Donut Chart

```python
df=pd.DataFrame(data=list(zip(l , s)) , columns=['Country' , 'Share'])
```
