```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,10)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1 ,0.1 , 0.8  ,0.8])
axes.plot(x, y, 'r')
```

```python
fig , axes = plt.subplots(nrows=1 , ncols=2)

axes[0].plot(x,y , 'r')
axes[0].set_title('1st plot')

axes[1].plot(y , x , 'b')
axes[1].set_title('2nd plot')

plt.tight_layout()
```

```python
axes
```

```python
fig , axes=plt.subplots(nrows=2 , ncols=2)
axes[0][0].plot(x,y , 'r')
axes[0][1].plot(y,x,'b')
axes[1,0].plot(x , y**2 , 'g')
axes[1,1].plot(x**3 , y , 'y')

plt.tight_layout()
```

```python
axes
```

```python
#Figure size and DPI
```

```python
fig=plt.figure(figsize=(8 , 2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x ,y)
```

```python
fig , axes=plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2 , ncols=1 , figsize=(8 , 5))
axes[0].plot(x,y,'r')
axes[1].plot(y,x ,'m')
```

```python
#Above figure will be saved
fig.savefig('My_pic.png' ,dpi=200 )
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2 , label='x**2')
axes.plot(x,x**3 , label='x**3')

axes.legend(loc=0)
```
