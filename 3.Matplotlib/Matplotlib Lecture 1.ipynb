```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0 , 5 , 11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x , y )
```

```python
plt.plot(x , y , 'r')
```

```python
plt.plot(x , y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('x vs x**2')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x , y , 'r')

plt.subplot(1,2,2)
plt.plot(y , x ,'b')
```

```python
plt.subplot(2 , 2 , 1)
plt.plot(x , y  , 'r')

plt.subplot(2 ,2 , 2)
plt.plot(y,x,'g')

plt.subplot(2,2,3)
plt.plot(x , y**2 , 'b')

plt.subplot(2,2,4)
plt.plot(x, y**3 , 'y')
```

```python
# Object Oriented Method:
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1 , 0.1 , 0.8 , 0.8])
axes.plot(x , y)

axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Demo Graph')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1 , 0.1 , 0.8 , 0.8])
axes2=fig.add_axes([0.2 , 0.5 , 0.2 , 0.3])

axes1.plot(x , y  , 'r')
axes1.set_title('Outer plot')
axes1.set_xlabel('Outer X')
axes1.set_ylabel('Outer Y')

axes2.plot(y , x , 'b')
axes2.set_title('inner plot')
axes2.set_xlabel('Inner X')
axes2.set_ylabel('Inner Y')
```
