```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,10)
y=x**2
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y , 'r' ,linewidth=5 , alpha=1 , ls='-' , marker='o' , markersize=10,
         markerfacecolor='y' , markeredgewidth=3 , markeredgecolor='black')
```
